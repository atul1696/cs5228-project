```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
palette = sns.husl_palette(10)

font = {'size'   : 14}
matplotlib.rc('font', **font)

from src.utils import read_csv
```

```python
auxSubzone, _ = read_csv('data/auxiliary-data/sg-subzones.csv')

auxInfraDict = {}
Infralist = ['sg-mrt-stations', 'sg-primary-schools', 'sg-secondary-schools', 'sg-shopping-malls', 'sg-commerical-centres']
for ele in Infralist:
    auxInfra, _ = read_csv('data/auxiliary-data/' + ele + '.csv')
    auxInfraDict[ele] = auxInfra
```

```python
label_dict = {'sg-mrt-stations': 'MRT \nStations',
              'sg-primary-schools': 'Primary \nSchools',
              'sg-secondary-schools': 'Secondary \nSchools',
              'sg-shopping-malls': 'Shopping \nMalls',
              'sg-commerical-centres': 'Commercial \nCentres'}
count_dict = {}
for ele in auxInfraDict:
    count_dict[label_dict[ele]] = len(auxInfraDict[ele])
```

```python
plt.figure(figsize=(7,4))
sns.barplot({"Landmarks": list(count_dict.keys()), "Count": list(count_dict.values())}, x="Landmarks", y="Count", palette=palette[:5], width=0.6).set(xlabel='Landmarks', ylabel='Count')

# plt.show()
plt.savefig("aux_data_number.pdf")
```

```python
for i, ele in enumerate(auxInfraDict):
    aux_data = auxInfraDict[ele]
    lat_arr, lng_arr = aux_data['lat'].astype(float).to_numpy(), aux_data['lng'].astype(float).to_numpy()
    
    plt.figure(figsize=(7,4))
    ax = sns.scatterplot({"Latitude": lat_arr, "Longitude" : lng_arr}, x="Longitude", y="Latitude", color=palette[i], s=150)
    ax.set_xlim(103.6,104.05)
    ax.set_ylim(1.21,1.49)
    
    plt.savefig("aux_data_" + ele + ".pdf")
```
